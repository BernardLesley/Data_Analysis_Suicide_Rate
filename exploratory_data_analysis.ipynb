```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

```python
master = pd.read_csv('master.csv')
master
```

```python
whr = pd.read_csv('world-happiness-report.csv')
```

```python
whr = whr.dropna()
whr
```

```python
df = pd.merge(master,whr,left_on=['country','year'],right_on=['Country name','year'])
```

```python
plt.figure(figsize=(10,6))
year = master.groupby(['year'])[['population','suicides_no']].sum().reset_index()
year['suicide_rate'] = year['suicides_no']*100000/year['population']
ax = sb.lineplot(x="year", y="suicide_rate",data=year)
ax.set(ylabel='Suicide Rate', xlabel='Year', title = 'Suicide Rate by Year across globe')
plt.show()
```

## Key Insight:
### Overall Suicide Trends Across the Globe

1. Suicide rates/100k pop peaked at 1995 (15.3)
2. This number has decreased steadily and linearly, the lowest being 11.5 in 2015 (about 25 percent decrease from the
   peak)

```python
plt.figure(figsize=(12,5))
year = master.groupby(['year','sex'])[['population','suicides_no']].sum().reset_index()
year['suicide_rate'] = year['suicides_no']*100000/year['population']
ax = sb.lineplot(x="year", y="suicide_rate",hue = 'sex',data=year,  legend="full", marker = 'o')
ax.set(ylabel='Suicide Rate', xlabel='Year', title = 'Suicide Trends Over Time (Gender)')
plt.show()
```

## Key Insight:
### Are men more inclined to commit suicide?

1. Globally, the rate of suicide for male has been consistently higher than that of female
2. The ratio between male suicide rates and female suicide rates is around 3.5, and this ratio remains relatively constant throughout the year

```python








dm = master.groupby(['country','year'])['suicides_no','population'].sum().reset_index()
dm['suicide_rate'] = dm['suicides_no']*100000/dm['population']
dm = dm[dm['suicides_no'].notnull()]
from sklearn.linear_model import LinearRegression
b = []
for country in dm.country.unique():
    tempdf = dm[dm['country'] == country]
    X = np.asarray(tempdf['year']).reshape([-1,1])
    y = np.asarray(tempdf['suicide_rate']).reshape([-1,1])
    model = LinearRegression()
    results = model.fit(X,y).coef_[0][0]
    b.append([country,results])

df = pd.DataFrame(b,columns=['country','factor']).sort_values(by='factor')
decrease = list(df['country'].head(10))
increase = list(df['country'].tail(10))
```

```python
plt.figure(figsize=(16,8))
leg = []
sb.set_palette("husl", 10)
for country in decrease:
    dm_country = dm[dm['country'] == country]
    ax = sb.lineplot(x="year", y="suicide_rate",data=dm_country, marker='o')
    leg.append(country)

plt.legend(title='Country', loc='upper left', labels=leg)
ax.set(ylabel='Suicide Rate', xlabel='Year', title = 'Top 10 Countries with highest decrease in suicide rates')
plt.show()
```

## Key Insight
### Decreasing Suicide Trend Among Nation

1. Estonia shows a quite significant decrease in the suicide rates, 1.31 less people (per 100k) commit suicide, the steepest decrease compared to other nation

```python


plt.figure(figsize=(16,8))
leg = []
sb.set_palette("husl", 10)
for country in increase:
    dm_country = dm[dm['country'] == country]
    ax = sb.lineplot(x="year", y="suicide_rate",data=dm_country, marker='o')
    leg.append(country)

plt.legend(title='Country', loc='best', labels=leg)
ax.set(ylabel='Suicide Rate', xlabel='Year', title = 'Top 10 Countries with hghest increase in suicide rates')
plt.show()
```

```python
by_gdp_country= df.groupby(['country']).agg({'population':'sum', 
                         'suicides_no':'sum', 
                         'Log GDP per capita':'mean',
                                                'Life Ladder' : 'mean',
                                        'Social support' : 'mean',
                                                'Perceptions of corruption':'mean',
                                               'Healthy life expectancy at birth' : 'mean',
                                                'Freedom to make life choices':'mean'}).reset_index()
by_gdp_country['suicide_rate'] = by_gdp_country['suicides_no']*100000/by_gdp_country['population']


rates = pd.DataFrame(by_gdp_country['suicide_rate'])
Life = pd.DataFrame(by_gdp_country['Life Ladder'])
gdp = pd.DataFrame(by_gdp_country['Log GDP per capita'])
soc_support = pd.DataFrame(by_gdp_country['Social support'])
corrupt = pd.DataFrame(by_gdp_country['Perceptions of corruption'])
exp = pd.DataFrame(by_gdp_country['Healthy life expectancy at birth'])
Freedom = pd.DataFrame(by_gdp_country['Freedom to make life choices'])
population = pd.DataFrame(by_gdp_country['population'])

linreg.fit(gdp, rates)
# Formula for the Regression line
regline_x = gdp
regline_y = linreg.intercept_ + linreg.coef_ * gdp

# Plot the Linear Regression line
f, axes = plt.subplots(6, 1, figsize=(24, 96))

axes[0].scatter(gdp, rates)
axes[0].plot(regline_x, regline_y, 'b-', linewidth = 3)
axes[0].set_xlabel("Log GDP/Capita", fontsize = 16)
axes[0].set_ylabel("Suicide Rates", fontsize = 16)
correlation = by_gdp_country['suicide_rate'].corr(by_gdp_country['Log GDP per capita'])
print("Log GDP per capita vs Suicide Rates")
print(correlation)
print("\n")


linreg.fit(Life, rates)
# Formula for the Regression line
regline_x = Life
regline_y = linreg.intercept_ + linreg.coef_ * Life
axes[1].scatter(Life, rates)
axes[1].plot(regline_x, regline_y, 'b-', linewidth = 3)
axes[1].set_xlabel("Life Ladder", fontsize = 16)
axes[1].set_ylabel("Suicide Rates", fontsize = 16)
correlation = by_gdp_country['suicide_rate'].corr(by_gdp_country['Life Ladder'])
print("Life ladder vs Suicide Rates")
print(correlation)
print("\n")

linreg.fit(soc_support, rates)
# Formula for the Regression line
regline_x = soc_support
regline_y = linreg.intercept_ + linreg.coef_ * soc_support
axes[2].scatter(soc_support, rates)
axes[2].plot(regline_x, regline_y, 'b-', linewidth = 3)
axes[2].set_xlabel("Social Support", fontsize = 16)
axes[2].set_ylabel("Suicide Rates", fontsize = 16)
correlation = by_gdp_country['suicide_rate'].corr(by_gdp_country['Social support'])
print("Social support vs Suicide Rates")
print(correlation)
print("\n")

linreg.fit(corrupt, rates)
# Formula for the Regression line
regline_x = corrupt
regline_y = linreg.intercept_ + linreg.coef_ * corrupt
axes[3].scatter(corrupt, rates)
axes[3].plot(regline_x, regline_y, 'b-', linewidth = 3)
axes[3].set_xlabel("Perception of corruption", fontsize = 16)
axes[3].set_ylabel("Suicide Rates", fontsize = 16)
correlation = by_gdp_country['suicide_rate'].corr(by_gdp_country['Perceptions of corruption'])
print("Perceptions of corruption vs Suicide Rates")
print(correlation)
print("\n")

linreg.fit(exp, rates)
# Formula for the Regression line
regline_x = exp
regline_y = linreg.intercept_ + linreg.coef_ * exp
axes[4].scatter(exp, rates)
axes[4].plot(regline_x, regline_y, 'b-', linewidth = 3)
axes[4].set_xlabel("Life Expectancy", fontsize = 16)
axes[4].set_ylabel("Suicide Rates", fontsize = 16)
correlation = by_gdp_country['suicide_rate'].corr(by_gdp_country['Healthy life expectancy at birth'])
print("Healthy life expectancy vs Suicide Rates")
print(correlation)
print("\n")


linreg.fit(Freedom, rates)
# Formula for the Regression line
regline_x = Freedom
regline_y = linreg.intercept_ + linreg.coef_ * Freedom
axes[5].scatter(Freedom, rates)
axes[5].plot(regline_x, regline_y, 'b-', linewidth = 3)
axes[5].set_xlabel("Freedom to make life choices", fontsize = 16)
axes[5].set_ylabel("Suicide Rates", fontsize = 16)
correlation = by_gdp_country['suicide_rate'].corr(by_gdp_country['Freedom to make life choices'])
print("Freedom to make life choices vs Suicide Rates")
print(correlation)
print("\n")



```

## Key Insight and conclusion

1. Individually, each of the variables that we analyze have low correlation with the suicide rates
2. Thus, it is not suitable to use only one variable and model it using linear regression
3. We can see that there is a clear distinction between sex and age group when plotted against the suicide rates across the year, it might be suitable to consider these two variables while building the model
