```python
import numpy as np
import pandas as pd
```

```python
master = pd.read_csv('master.csv')
master
```

```python
whr = pd.read_csv('world-happiness-report.csv')
```

```python
whr = whr.dropna()
whr
```

```python
df = pd.merge(master,whr,left_on=['country','year'],right_on=['Country name','year'])
```

```python
df
```

```python
df.info()
```

```python
df=df.drop(df.columns[[0,1,4,5,7,8,9,10,11,12,20,21]], axis=1)
```

```python
df
```

```python
df = pd.get_dummies(df, columns=['sex','age'])
```
